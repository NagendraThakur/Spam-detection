```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.model_selection import train_test_split
```

```python
dataset_file = open("spam_dataset.txt", encoding="utf8")
data = dataset_file.read().split("\n")    
data = [d.split(",",1) for d in data]
data
```

```python
data = pd.DataFrame(data)
data.columns = ["label","text"]
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.text, data.label, test_size=0.2) # 80% training and 20% test
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
train_data = {'text' : X_train, 'label': y_train}
train_data = pd.DataFrame(data=train_data)
train_data
```

```python
test_data = {'text' : X_test, 'label': y_test}
test_data = pd.DataFrame(data=test_data)
test_data
```

```python
vocab = []
spam_data = []
ham_data = []
spam_count=0
ham_count=0
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))
for index, category in train_data.iterrows():
    each_text = []
    word_tokens = word_tokenize(category['text'].lower().replace(".","").replace("<","").replace(">","").replace(",","").replace("+","").replace("!","").replace("?","").replace('"','').replace("(","").replace(")","").replace("'",""))
    for w in word_tokens:
        if w not in stop_words:
            w = ps.stem(w)
            each_text.extend([w])
            vocab.extend([w])
    if (category['label'] == 'spam'):
        spam_count+= 1
        spam_data.append(each_text)
    else:
        ham_data.append(each_text)
        ham_count+= 1
```

```python
ham = []
for text in ham_data:
    ham_vector = []
    for word in vocab:
        ham_vector.append(text.count(word))
    ham.append(ham_vector)
```

```python
def mail(new_mail):
    new = [new_mail]
    new_words= []
    for each in new:
    #     print(each)
        word_tokens = word_tokenize(each.lower().replace(".","").replace(",","").replace("<","").replace(">","").replace("+","").replace("!","").replace("?","").replace('"','').replace("(","").replace(")","").replace("'",""))
    #     print (word_tokens)
        for w in word_tokens:
             if w not in stop_words:
                 new_words.append(ps.stem(w))

#     new_words=[new_words]
#     new_words=new_words[0]
    return(new_words)
    print(new_words)
    
```

```python
def predict(words,probability_spam,probability_ham,probabilities_spam,probabilities_ham,vocab):
    spam = probability_spam
    ham = probability_ham
    for word in words:
        try:
            word_probability = probabilities_spam[vocab.index(word)]
        except:
            word_probability = 1
        finally:
            spam *= word_probability
            ham *= word_probability
#     print(spam,ham)
    if spam > ham:
        return "spam"
#         return " ".join(words),"spam"
    else:
        return "ham"
#         return " ".join(words),"ham"
```

```python
accurate_result=0
for index, category in test_data.iterrows():
    accuracy_label = predict(mail(category['text']),probability_spam,probability_ham,probabilities_spam,probabilities_ham,vocab)
    if (category['label'] == accuracy_label):
        accurate_result+= 1

print(accurate_result)
print(len(test_data))
accuracy=accurate_result/len(test_data)
print('The accuracy of the system is',accuracy)
```
